```python
import numpy as np
import arviz as az
from branching_sus.implementation import SubsetSimulation
import branching_sus.plot as bplt
import branching_sus.estimate as est
import matplotlib.pyplot as plt
az.style.use('arviz-darkgrid')
```

```python
def performance_function(x):
    return x[0] + x[1]

subset_simulation = SubsetSimulation(performance_function=performance_function,
                                     dimension=2,
                                     level_size=1000,
                                     threshold=9,
                                     level_probability=0.1,
                                     seed=0,
                                     verbose=True
                                    )

subset_simulation.run()
```

```python
bplt.all_levels(subset_simulation)
```

```python
est.exceedance_probability(subset_simulation,9)
```

```python
est.cov(subset_simulation,9)
```

```python
bplt.ccdf(subset_simulation,9,100)
```
